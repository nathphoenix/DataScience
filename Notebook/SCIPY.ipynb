# How to sample from  general multivariate distribution

```python
from scipy.stats import multivariate_normal as mvn
cov = np.array([[1, 0.8], [0.8, 3]])
mu = np.array([0, 2]) #where two is the mean
#this means we have a covariance of 1 in the first dimension and variance of 3 in the second dimension
#and the covariance between the two dimension is 0.8
r = mvn.rvs(mean=mu, cov=cov, size=1000)
plt.scatter(r[:,0], r[:,1])
plt.axis("equal")
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000) #usimg numpy to achieve the above result
plt.scatter(r[:,0], r[:,1])
plt.axis("equal")
plt.show()
```
